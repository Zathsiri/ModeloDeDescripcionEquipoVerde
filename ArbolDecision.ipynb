```python
#Importamos las librerias
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy.stats import kstest
from scipy.stats import anderson
import numpy as np
import seaborn as sns
import fitter
from fitter import Fitter, get_common_distributions, get_distributions
from scipy.stats import beta
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier

```

```python
#creacion del dataframe
import io
from google.colab import files

uploaded = files.upload() 
df = pd.read_csv(io.BytesIO(uploaded['Fertile_Man_WHO_2020.csv']))
df
```

```python
df = df.drop(columns = 'Vitality')
```

```python
#probamos cuando tenemos total_motility y no contamos con Immotile_Spermatoza
df['Immotile_Spermatozoa'] = df['Immotile_Spermatozoa'].fillna(100-df['Total_Motility'])
#probamos cuando tenemos Immotile_Spermatoza y no contamos con total_motility. Despejando la formula
df['Total_Motility'] = df['Total_Motility'].fillna(100-df['Immotile_Spermatozoa'])
df
```

```python
df['Progressive_Motility'] = df['Progressive_Motility'].fillna(df['Total_Motility']-df['Non-progressive_Motility'])
df['Non-progressive_Motility'] = df['Non-progressive_Motility'].fillna(df['Total_Motility']-df['Progressive_Motility'])
```

```python
df=df.dropna()
df.info()
```

Encontrar los valores para cada una de las clases de fertilidad que existe

# 0 - infertil 
1 - medio
## 2 - super fertil

```python
#Dataframe completo para saber si se ejecuto bien la condicion
outcome = []
normal_form = []
semen_Volume = []
#Contadores para cada tipo
count_low = 0
count_normal = 0
count_extra = 0

#Listas para dataframes de cada tipo
nf_low = []
nf_normal = []
nf_extra = []


for i in range(len(df)):
  if df.iloc[i]['Normal_Forms'] < 4.0:
    fertility_impared = 0
    nf_low.append(df.iloc[i]['Normal_Forms'])
    outcome.append(fertility_impared)
    count_low = count_low + 1
  elif df.iloc[i]['Normal_Forms'] > 14.0:
    fertility_extra = 2
    nf_extra.append(df.iloc[i]['Normal_Forms'])
    outcome.append(fertility_extra)
    count_extra = count_extra + 1
  else:
    fertility_normal = 1
    nf_normal.append(df.iloc[i]['Normal_Forms'])
    outcome.append(fertility_normal)
    count_normal = count_normal + 1
  normal_form.append(df.iloc[i]['Normal_Forms'])
```

```python
#Cuantos valores hay de cada uno
print("Low: ",count_low)
print("Normal: ", count_normal)
print("Extra: ",count_extra)
```

```python
#agregar el outcome
df['Outcome'] = outcome
```

```python
#Partir dependiedo de la categoria

df_low = df[df['Outcome'] == 0]
df_normal = df[df['Outcome'] == 1]
df_extra= df[df['Outcome'] == 2]
```

```python
X = df_ready_all.drop(columns = ['Publisher','Total_Motility','Progressive_Motility','Non-progressive_Motility','Normal_Forms','Outcome'])
y = df_ready_all['Outcome']
X_train, X_test, y_train, y_test = train_test_split(
                     X, y, test_size=0.30)
```

```python
X
```

```python
#entrenamiento
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
```

```python
#Prediccion
y_pred = clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
